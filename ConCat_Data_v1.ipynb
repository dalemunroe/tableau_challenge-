```python
import pandas as pd
import random
```

```python
# import raw data file

df_1805_raw = pd.read_csv("raw_data/201805-citibike-tripdata.csv", low_memory=False)
df_1805_raw.head(1)
```

```python
# reduce raw dataframe doen to match 2022 and 2022 dataframe

df_1805_reduced = df_1805_raw[["usertype",
                            "start station id",
                            "start station name",
                            "end station id",
                            "end station name",
                            "start station latitude",
                            "start station longitude",
                            "end station latitude",
                            "end station longitude",
                            "starttime",
                            "stoptime",
                            ]]
```

```python
# reorder, remane headings, and format columns in dataframe.

df_1805_cleaned = df_1805_reduced.rename(columns={"usertype" : "member_casual",
                    "start station id" : "start_station_id", 
                    "start station name" : "start_station_name", 
                    "end station id" : "end_station_id", 
                    "end station name" : "end_station_name",
                    "start station latitude" : "start_lat", 
                    "start station longitude" : "start_lng",
                    "end station latitude" : "end_lat",
                    "end station longitude" : "end_lng",
                    "starttime" : "started_at",
                    "stoptime" : "ended_at",})
df_1805_cleaned["started_at"]=pd.to_datetime(df_1805_cleaned["started_at"])
df_1805_cleaned["ended_at"]=pd.to_datetime(df_1805_cleaned["ended_at"])
df_1805_cleaned.info()
```

```python
# Loop through the *.csv files and take random samples (10%) and append to a combined dataframe

csv_list = ["cleaned201805.csv","cleaned201905.csv", "cleaned202005.csv", "cleaned202105.csv", "cleaned202205.csv"]
combined_data = pd.DataFrame()
p = 0.1
for x in csv_list:
    data=pd.read_csv(x, low_memory=False, header = 0, skiprows = lambda i: i>0 and random.random()>p)
    combined_data=combined_data.append(data,ignore_index=True)
```

```python
# Format datatime columns.

combined_data["started_at"]=pd.to_datetime(combined_data["started_at"])
combined_data["ended_at"]=pd.to_datetime(combined_data["ended_at"])
combined_data.info()
```
